```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../../data/LABELLED_DATASET.csv")
df.head()
```

```python
language_constructors = {}
for index, row in df.iterrows():
    constructors = row['kind_conflict']
    constructors = constructors.split(',')
    decision = row['developerdecision']
    for constructor in constructors:
        constructor = constructor.strip()
        if constructor in language_constructors:
            line = language_constructors[constructor]
            if decision in line:
                language_constructors[constructor][decision]+=1
            else:
                language_constructors[constructor][decision]=1
        else:
            language_constructors[constructor] = {}
            language_constructors[constructor][decision]=1
print(f"Total numbers of chunks in the dataset: {len(df)}")
print("Number of times each constructor appears in the dataset considering all chunks:")
# print(language_constructors)
result = pd.DataFrame(language_constructors)
```

```python
result = result.transpose()
```

```python
# result.loc['Total',:]= result.sum(axis=0)
result.loc[:,'Total'] = result.sum(axis=1)
result = result.convert_dtypes()
result = result.drop(columns = ['UnknownConcatenation'])
```

```python
result.sort_values(by=['Total'], ascending=False)
```

```python
columns = list(result.columns)
columns.remove("Total")
data = []
indexes = []
for index, row in result.iterrows():
    new_row_values = []
    indexes.append(index)
    for column in columns:
        confidence_value = row[column]/row["Total"]
        new_row_values.append(confidence_value)
    data.append(new_row_values)

confidence = pd.DataFrame(data, columns=columns, index=indexes)
print("Confidence Language_construct => developer_decision")
print("Confidence A=>B: number of transactions that include both A and B divided by the number of transactions that include B")
confidence
        
```

```python
consequents = list(df['developerdecision'].unique())
support = {}
for consequent in consequents:
    consequent_support = len(df[df['developerdecision']==consequent])/len(df)
    support[consequent] = consequent_support
    
print("Support of a given developer_decision:")
print("Probability of the developer_decision over all the transactions")
support
```

```python
data = []
indexes = []
for index, row in result.iterrows():
    new_row_values = []
    indexes.append(index)
    for column in columns:
        confidence_value = confidence.loc[index][column]
        support_value = support[column]
        lift_value = confidence_value/support_value
        new_row_values.append(lift_value)
    data.append(new_row_values)

lift = pd.DataFrame(data, columns=columns, index=indexes)
print("Lift language_constructor => developer_decision")
print("Calculated as confidence(language_constructor => developer_decision) divided by the support(developerdecision)")
lift
        
```

```python
rules_increase = {}
rules_decrease = {}
threshold = 0.5
for index, row in lift.iterrows():
    for column in lift.columns:
        if not pd.isnull(row[column]):
            if row[column] >= 1 + threshold:
                rule = f'{index} => {column}'
                rules_increase[row[column]] = rule
            if row[column] <= 1 - threshold:
                rule = f'{index} => {column}'
                rules_decrease[row[column]] = rule
```

```python
data = []
print(f"Mined rules with at least {threshold*100:.2f}% increased chance")
for key in sorted(rules_increase, reverse=True):
    data.append([rules_increase[key], key])
pd.DataFrame(data, columns=['Rule', "Lift"])
    
```

```python
data = []
print(f"Mined rules with at least {threshold*100:.2f}% decreased chance")
for key in sorted(rules_decrease):
    data.append([rules_decrease[key], key])
pd.DataFrame(data, columns=['Rule', "Lift"])
```
