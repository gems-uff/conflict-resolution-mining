```python
import pandas as pd
```

```python
df = pd.read_csv("../../data/LABELLED_DATASET.csv")
df.head()
```

```python
language_constructors = {}
for index, row in df.iterrows():
    constructors = row['kind_conflict']
    constructors = constructors.split(',')
    decision = row['developerdecision']
    for constructor in constructors:
        constructor = constructor.strip()
        if constructor in language_constructors:
            line = language_constructors[constructor]
            if decision in line:
                language_constructors[constructor][decision]+=1
            else:
                language_constructors[constructor][decision]=1
        else:
            language_constructors[constructor] = {}
            language_constructors[constructor][decision]=1
print(f"Total numbers of chunks in the dataset: {len(df)}")
print("Number of times each constructor appears in the dataset considering all chunks:")
# print(language_constructors)
result = pd.DataFrame(language_constructors)
```

```python
result = result.transpose()
```

```python
# result.loc['Total',:]= result.sum(axis=0)
result.loc[:,'Total'] = result.sum(axis=1)
result = result.convert_dtypes()
```
